## Simulation Results
Here we present numerical simulations of a group of $N = 6$ robots manipulating a rigid body in $SE(3).$ We include implementations of both the typical adaptation law & one with l-p regularization.

### Physical Parameters & Constants

```julia
#Physical parameters
rho = 100 #density, kg/m^3 
r = 0.5 #cylinder radius, m
h = 0.2 #cylinder height, m
m = rho*pi*(r^2)*h #cylinder mass, kg



r_p = [0,0,h/2] #measurement point
N = 3
# r_i = [[0,-1,1],[1,1,0],[0,1,-1],[1,2,1]]
# r_i = [[1,1,0],[0,-1,1],[0,1,-1],[1,2,1]]
r_i = [randn(3) for i = 1:N] #initial r_i

#compute inertia tensor
I_cm = (1/12)*Diagonal([m*(3*r^2+h^2),m*(3*r^2+h^2),6*m*r^2])
I_p = I_cm + m*(r_p'*r_p*I -r_p*r_p')


#frequencies & phase shifts for reference signal computation
num_freq = 15
ratio_l = 15
ratio_r = 10
freqs = rand(6,num_freq)
# phases = zeros(6,num_freq)
phases = 5*rand(6,num_freq)

#define  reference signals (combinations of sinusoids)
function w_d(t)
    w_dt = (ratio_r/length(freqs))*sum(vcat(cos.(freqs[1,:]*t.+phases[1,:])',sin.(freqs[2,:]*t.+phases[2,:])',cos.(freqs[3,:]*t.+phases[3,:])'),dims=2)[:]
    if dimension == 2
        w_dt[1] = 0
        w_dt[2] = 0
    end
    return w_dt
end

function al_d(t)
    al_dt = (ratio_r/length(freqs))*sum(vcat(-freqs[1,:]'.*sin.(freqs[1,:]*t.+phases[1,:])',freqs[2,:]'.*cos.(freqs[2,:]*t.+phases[2,:])',-freqs[3,:]'.*sin.(freqs[3,:]*t.+phases[3,:])'),dims=2)[:]
    if dimension == 2
        al_dt[1] = 0
        al_dt[2] = 0
    end
    return al_dt
end

function v_d(t)
    v_dt = (ratio_l/length(freqs))*sum(vcat(cos.(freqs[4,:]*t.+phases[4,:])',cos.(freqs[5,:]*t.+phases[5,:])',cos.(freqs[6,:]*t.+phases[6,:])'),dims=2)[:]
    if dimension == 2
        v_dt[3] = 0
    end
   return v_dt
end

function a_d(t)
    a_dt = (ratio_l/length(freqs))*sum(vcat(-freqs[4,:]'.*sin.(freqs[4,:]*t.+phases[4,:])',-freqs[5,:]'.*sin.(freqs[5,:]*t.+phases[5,:])',-freqs[6,:]'.*sin.(freqs[6,:]*t.+phases[6,:])'),dims=2)[:]
    if dimension == 2
        a_dt[3] = 0
    end
    return a_dt
end

function dq_des(t)
   return vcat(v_d(t),w_d(t)) 
end

function ddq_des(t)
   return vcat(a_d(t),al_d(t))
end
```

### Gain

```julia
# ICL gain
N_o = 20
ICL_update = false
ICL_switcher = true
k_cl_m = 0.01
k_cl_rp = 0.2
k_cl_I = 0.2
k_cl_Izz = 0.3
k_cl_ri = 0.6 #0.02
k_cl_gain = Diagonal(vcat(k_cl_m,k_cl_rp*ones(3),k_cl_I*ones(5),k_cl_Izz*ones(1),k_cl_ri*ones(3*N)))

# gamma o
g_m = 7
g_rp = 6
g_I = 1.0
g_Izz = 3
g_ri = 0.25 #0.2 1.5
g_o = Diagonal(vcat(g_m,g_rp*ones(3),g_I*ones(5),g_Izz*ones(1),g_ri*ones(3*N)))

#Kd matrix
Kdl = 7 #3.5
Kdr = 3 #1
Kd = Diagonal(vcat(Kdl*ones(3),Kdr*ones(3)))

#sliding parameter
lambda = 1.5

#deadband in which to stop adaptation
deadband = 0.001

#simulation parameters
dt = 0.01 #timestep -- numerical integration via Heun's; error is ~ O(dt^2) 0.005 0.02
T = 100 #final time
t_drop = T+10 #time for drop -- use t_drop < T for agents to turn off
dimension = 3 # n = 2 or n = 3

# addition random force
rand_f = false

# set constraint
force_bound = 40
```

### Initialize

```julia
##setup constants, etc.
# Random.seed!(318) #seed rand for repeatability

# a_hat = [0.6*randn(10+3*N) for i = 1:N] #initial a_hat 1.2
# a_hat = [zeros(10+3*N) for i = 1:N]
a_hat = zeros(10+3*N)

a_hat[11:10+3*N] = 0.3*randn(3*N)

# for i=1:N
#     a_hat[i][11:13] = r_i[i]
# end

# for i=1:N
#     a_hat[i][11:13] = 0.8*randn(3,1) #0.8
# end

# for i=1:N
#     a_hat[i][11:13] = 1.0*zeros(3,1)
# end

#true a_i
a_I = [I_p[1,1], I_p[1,2], I_p[1,3], I_p[2,2], I_p[2,3], I_p[3,3]]
a = vcat(m,m*r_p,a_I)
for i = 1:N
    a = vcat(a,r_i[i])
end
a_i = a

#sample random initial orientations
q = rand(4)
q = q/norm(q) #random quaternion
q_d = rand(4)
q_d = q_d/norm(q_d) #random desired quaternion
#cast both to rotation matrices
R = quat_to_rot(q)
R_d = quat_to_rot(q_d)

dq = zeros(6) #initial velocity = 0

#inital state
w = zeros(3)
x = zeros(3)
x_d = zeros(3)
dx = zeros(3)

#inital force
force = [zeros(3, 1) for i = 1:N]
F = [zeros(6) for i = 1:N]



p = 2 #l-p norm to be used for regularization;
#note: we only implement l1, l2, since physical scaling issues => numerical problems for p > 2.


P_o = 3e1*inv(Diagonal(ones(10+3*N)).+1e-2)
P_r = 3e1*inv(Diagonal(ones(3)))
tol=1e-5


#storage for sim. data
s_t = []
rot_err = []
V = []
s = zeros(6)
lyap_params = []
err_mean = []
arm_est = []
dV_params = []
true_tau = []

index = 1
mat_matrix = zeros(10+3*N, N_o)
dx_last = zeros(3)
w_last = zeros(3)
rand_force = []

true_tau_integral = (6,0)
force_debug = zeros(0,6)

param_est_sum = zeros(0, 10+3*N)
param_est_err = zeros(0, 10)

arm_est_sum = zeros(0, 3)
arm_est_err = zeros(0, N)

res_m = zeros(0, N)
res_x = zeros(0, N)
res_x_err = zeros(0, N)
central_arm = zeros(0, 3*N)

yo_rank = zeros(0, 3*N)



global H = vcat(hcat(m*I,m*skew(R*r_p)),
         hcat(-m*skew(R*r_p), R*I_p*R'))
verbose = false
dropped = false
switch_apply = 1
condition = 1
switch_time = 1

split_1 = Diagonal(rand(Uniform(0.40,0.60), 6))
split_2 = Diagonal(vcat(ones(6))) - split_1
```

### Dynamics Implementation

```julia
function adaptive_dyn(x,x_d,R,R_d,dq,a_hat,t;t_drop=Inf)
    #inputs: position x, desired pos. x_d, orientation R, des. orient. R_d, velocity dq, param estimates a_hat, r_hat, time t;
    #optional : drop time t_drop
    #outputs: state & parameter derivatives

    
    if t%switch_time == 0
        global condition = (condition + 1) % N
    end
    
    #unpack variables
    dx = deepcopy(dq[1:3])
    w = deepcopy(dq[4:6])
    
    #compute error signals
    Re = R_d'*R
    we = w - w_d(t)
    
    sigma = we + lambda*R_d*vee(Pa(Re))
    dx_t = dx - v_d(t)
    x_t = x - x_d
    
    s = vcat(dx_t+lambda*x_t,sigma)
    
    al_r = al_d(t) - lambda*skew(w_d(t))*R_d*vee(Pa(Re)) - lambda*R_d*vee(Pa(skew(R_d'*we)*Re))
    w_r = w_d(t) - lambda*R_d*vee(Pa(Re))
    a_r = a_d(t) - lambda*dx_t
    v_r = v_d(t) - lambda*x_t
     
    ddq_r = vcat(a_r, al_r)
    dq_r = vcat(v_r, w_r)
    
    #compute regressors


    Y_l = hcat(a_r,-skew(al_r)*R-skew(w)*skew(w_r)*R,zeros(3,6+3*N))
    Y_r = hcat(zeros(3,1),skew(a_r)*R+skew(w)*skew(v_r)*R-skew(w_r)*skew(dx)*R,
        R*L(R'*al_r)+skew(w)*R*L(R'*w_r))
    for i=1:N
        Y_r = hcat(Y_r,skew(force[i])*R)
    end
    Y_o = vcat(Y_l,Y_r)


    
    
    #compute true dynamics matrices
    H = vcat(hcat(m*I,m*skew(R*r_p)),
         hcat(-m*skew(R*r_p), R*I_p*R'))
        
    C = vcat(hcat(zeros(3,3),m*skew(w)*skew(R*r_p)),
        hcat(-m*skew(w)*skew(R*r_p),skew(w)*R*I_p*R'-m*skew(skew(R*r_p)*dx)))
    
    off_diag = m*skew(w)*R*skew(r_p)*R' - m*R*skew(r_p)*R'*skew(w)
    H_dot = vcat(hcat(zeros(3,3),off_diag),
                    hcat(-off_diag,skew(w)*R*I_p*R'-R*I_p*R'*skew(w)))



    ### ICL
    y_o_cl_integral = []
    if ICL_update && ICL_switcher
        y_l = hcat(dx-dx_last, -skew(w-w_last)*R-(skew(w)*skew(w)*R)*dt, zeros(3,6+3*N))
        y_r = hcat(zeros(3,1), skew(dx-dx_last)*R, R*L(R'*(w-w_last))+(skew(w)*R*L(R'*w))*dt)
        for i = 1:N
            y_r = hcat(y_r, skew(force[i])*R*dt)
        end
        y_o_cl_integral = vcat(y_l, y_r)

        global yo_rank = vcat(yo_rank, y_o_cl_integral[:, 11:end])
        
        if index > N_o
            global index = 1
        end

        global true_tau_integral = zeros(6, 1)
        for i = 1:N
            true_tau_integral = true_tau_integral + true_tau[i] * dt
        end
        y_o_cl_integral_a_hat = y_o_cl_integral * a_hat
        mat_now = y_o_cl_integral'*(y_o_cl_integral_a_hat - true_tau_integral)

        global mat_matrix[:, index] = mat_now[:]

        mat_sum = sum(mat_matrix, dims=2)

        global index = index + 1
        
        global dx_last = dx
        global w_last = w

    end

    

    ###

    
    #apply control law
    input = zeros(6)
    global true_tau = []
    global force = []
    
    

#     global F[switch_apply+1] = Y_o * a_hat[switch_apply+1] / N - Kd*s #implement one control law rand(Uniform(0.8,1.2), 6,6) * 
    
    for i = 1:N
        global F[i] = -Kd*s
    end
    
#     if switch_apply+1 == 1
#         F[switch_apply+1] += Y_o * a_hat[switch_apply+1]
#     end

    

    F[condition+1] += Y_o * a_hat




#     if switch_apply+1 == 1
#         F[switch_apply+1] = Y_o * a_hat[switch_apply+1]
#     elseif switch_apply+1 == 2
#         F[switch_apply+1] = -3*split_1*Kd*s
#     elseif switch_apply+1 == 3
#         F[switch_apply+1] = -3*split_2*Kd*s
#     end


#     for i=1:6
#         if F[switch_apply+1][i] > force_bound
#             F[switch_apply+1][i] = force_bound
#         elseif F[switch_apply+1][i] < -force_bound
#             F[switch_apply+1][i] = -force_bound
#         end
#     end

    for i = 1:N
        tau = F[i]  #compensate for est. torque

        G = vcat(hcat(I,zeros(3,3)),hcat(skew(R*r_i[i]),I))
        input += G*tau #compute real applied wrench; add to running

        push!(true_tau, tau)
        push!(force,tau[1:3])
    end

    global force_debug = vcat(force_debug, input')



    #use root finding to solve for accelerations (avoid inverting H)
    f(ddq) = H*ddq + C*dq - input
    ddq = nlsolve(f,zeros(6),autodiff=:forward).zero
    
    #implement adaptation laws
    Y_g, dr, a_t, r_t = [], [], [], []
    da, g_r = [], [], []
    

    ### ICL
    if ICL_update && ICL_switcher
        ri_switch = vcat(ones(10),zeros(3*N))
        ri_switch[8 + (condition+1) * 3 : 10 + 3 * (condition+1)] = ones(3)
        ri_switch_go = Diagonal(ri_switch)
        da = -g_o*Y_o'*s - k_cl_gain*g_o*ri_switch_go*reshape(mat_sum[:, 1], 10+3*N)
    ###
    else
        da = -g_o*Y_o'*s
    end
    a_t = a_hat - a_i

    
    global ICL_update = true
    
    ds = ddq - ddq_r
    ds2 = vcat(ddq[1:3]-a_d(t)+lambda*(dx-v_d(t)),
        ddq[4:6]-al_d(t)+lambda*(skew(w_d(t))*R_d*vee(Pa(Re)) + R_d*vee(Pa(skew(R_d'*we)*Re))))
    
    dV = s'*(-H*ddq_r - C*dq_r)  #s'(-Y_oo_i)
    dV2 = s'*H*ds + (1/2)*s'*H_dot*s
    r_err = 0
    o_err = 0
    
    for i = 1:N
        dV += s'*(F[i])+ a_t'*inv(g_o)*da
        dV2 += a_t'*inv(g_o)*da
    end
    
    if verbose
        @show norm(H_dot-2*C+(H_dot-2*C)')/norm(H_dot-2*C) #check H_dot-2*C s.s.
        @show dV #check dV <= 0
        @show dV - dV2 #check dV equal when computed 2 different ways
        @show dV + N*s'*Kd*s #check dV = -N*s'*Kd*s
        @show norm(ds-ds2) #check ds equal when computed 2 different ways
    end
    
    #output accelerations & derivatives
    return ddq, da, dr, dV, ds
end
```

### Run Simulations

```julia
##############################
###### Simulation Start ######
##############################
for t in 0:dt:T
    if t >= t_drop && !dropped
        N = convert(Int,N/2) # ensure that N is a integer
        dropped = true
    end
    
    #unpack variables
    dx = deepcopy(dq[1:3])
    w = deepcopy(dq[4:6])

    #compute error signals
    Re = R_d'*R
    we = w - w_d(t)
    sigma = we + lambda*R_d*vee(Pa(Re))
    dx_t = dx - v_d(t)
    x_t = x - x_d
    s = vcat(dx_t+lambda*x_t,sigma)
    
#     check if simulation has diverged -- if so, decrease dt
    if norm(s) > 1e3
        @warn "diverged"
        break
    end
    
    if t > 20 && norm(s) > 10
        print(t)
        @warn "divergedd"
        break
    end
    
    #dynamics matrices (for debug)
    H = vcat(hcat(m*I,m*skew(R*r_p)),
         hcat(-m*skew(R*r_p), R*I_p*R'))
    
    C = vcat(hcat(zeros(3,3),m*skew(w)*skew(R*r_p)),
        hcat(-m*skew(w)*skew(R*r_p),skew(w)*R*I_p*R'-m*skew(skew(R*r_p)*dx)))
    ## for H dot
    off_diag = m*skew(w)*R*skew(r_p)*R' - m*R*skew(r_p)*R'*skew(w)
    H_dot = vcat(hcat(zeros(3,3),off_diag),
                    hcat(-off_diag,skew(w)*R*I_p*R'-R*I_p*R'*skew(w)))
    
    #compute Lyap. function & param errors


    V_curr = 0.5*(s'*H*s)


    a_t = a_hat-a_i
    V_curr += pnorm_divergence(P_o*a_i,P_o*a_hat,p,tol)

    
    push!(V,V_curr)

    at_mean = (1/N)*sum(a_t)
    
    #logging
    push!(s_t,norm(s))
    push!(rot_err,norm(R_d'*R-I))
    
    
    ######################
    ###### Dynamics ######
    ######################
    #get dynamics @ current time
    global rand_force = []
    ddq, da, dr, dV, ds = adaptive_dyn(x,x_d,R,R_d,dq,a_hat,t)
        

    if norm(s) > deadband #if above deadband, update parameters
        a_hat = a_hat + N*dt*da
    else
        a_hat = a_hat
    end

    
    R_d = exp(dt*skew(w_d(t)))*R_d
    R = exp(dt*skew(w))*R
    
    x_d = x_d + dt*v_d(t)
    x = x + dt*dx
    dq = dq + dt*ddq
    
    dx = deepcopy(dq[1:3])
    w = deepcopy(dq[4:6])

    param_est_sum = vcat(param_est_sum, a_hat')
    param_est_err = vcat(param_est_err, a_hat[1:10]' - a[1:10]')
    
    # for only UAV1
#     param_est_sum = vcat(param_est_sum, transpose(a_hat[1]))
#     param_est_err = vcat(param_est_err, transpose(a_hat[1][1:10]) - a[1:10]')
    
    central_arm = vcat(central_arm, hcat([a_hat[8+3*i:10+3*i]' - r_i[i]' for i = 1:N]...))
   

    
#     param_est_sum = vcat(param_est_sum, transpose(a_hat))
#     param_est_err = vcat(param_est_err, transpose(a_hat[1:10])-a[1:10]')
    
#     res_m = vcat(res_m, hcat([a_hat[i][1] for i = 1:N]...))
#     res_x = vcat(res_x, hcat([a_hat[i][11] for i = 1:N]...))
#     res_x_err = vcat(res_x_err, hcat([a_hat[1][10+i] - r_i[1][i] for i = 1:3]...))
    
#     arm_est_sum = vcat(arm_est_sum, sum(transpose(hcat([r_hat[i] for i = 1:N]...)), dims = 1))

    arm_est_err = vcat(arm_est_err, hcat([norm(a_hat[8+3*i:10+3*i]-r_i[i]) for i = 1:N]...))
end


s_l2 = s_t;
rot_l2 = rot_err;
lyap_l2 = V;
arm_l2 = arm_est;
```
